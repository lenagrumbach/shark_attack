## Parte 2 - Identificação das colunas que contém informações relevantes para esta análise e limpeza dos dados.

```python
df.head()
```

###2.1 Analisando os países (coluna 'Country').

```python
(df['Country'].unique())
```

```python
df['pais'] = df['Country'].str.upper()
```

**Os 10 países onde mais aconteceram ataques de tubarão.**

```python
total_por_pais = ((df['pais'].value_counts(normalize=True).round(2)*100).astype(str)+' %').to_frame().reset_index().rename(columns = {'pais':'percentual_ataque', 'index':'pais'})
total_por_pais.head(10)
```

###2.2 Analisando as localidades (coluna 'Area').

```python
(df['Area'].unique())
```

```python
df['localidade'] = df['Area'].str.title()
```

```python
for nome in total_por_pais['pais']:
  mascara_pais = total_por_pais['pais'] == nome
  total_por_pais.loc[mascara_pais, 'localidades_mais_perigosas'] = str(df.query(f'pais == "{nome}"').groupby('localidade')['Case Number'].count().sort_values(ascending=False)[:3].index.to_list())
```

**As 3 localidades em cada pais onde mais ataques ocorreram.**

```python
total_por_pais[:10].style.set_properties(subset=['localidades_mais_perigosas'], **{'width': '300px'})
```

###2.3 Analisando as atividades (coluna 'Activity').

```python
df['Activity'].value_counts()[:30]
```

```python
df['atividade'] = df['Activity'].str.lower().str.replace('.*surf.*', 'Surf', regex=True).str.replace('.*swimming.*', 'Nado', regex=True).str.replace('.*fishing.*', 'Pesca', regex=True).str.replace('.*diving.*|.*snorkeling.*', 'Mergulho', regex=True).str.replace('.*bathing.*|.*wading.*|.*standing.*|.*treading water.*', 'Estar na agua', regex=True)
```

**As 4 atividades aquáticas mais perigosas.**

```python
atividades = df['atividade'].value_counts().to_frame().reset_index().rename(columns = {'index':'atividade', 'atividade':'ataques'})
atividades[:4]
```

###2.4 Analisando os meses (coluna 'Case Number').



```python
df['Case Number'][:20]
```

```python
df['mes'] = df['Case Number'].str.replace('.*\d{4}\.(\d{2}).*', r'\1', regex=True)
df['mes'] = df['mes'].str.replace('01', 'Jan', regex=True).str.replace('02', 'Fev', regex=True).str.replace('03', 'Mar', regex=True).str.replace('04', 'Abr', regex=True).str.replace('05', 'Mai', regex=True).str.replace('06', 'Jun', regex=True).str.replace('07', 'Jul', regex=True).str.replace('08', 'Ago', regex=True).str.replace('09', 'Set', regex=True).str.replace('10', 'Out', regex=True).str.replace('11', 'Nov', regex=True).str.replace('12', 'Dez', regex=True)#.value_counts()
```

```python
ataques_por_mes = df['mes'].value_counts().to_frame().reset_index().rename(columns = {'index':'mes', 'mes':'ataques'})
```

**Os 6 meses em que tiveram mais ataques ao longo dos anos.**

```python
ataques_por_mes[:6]
```

###2.5 Analisando os anos (coluna 'Case Number').

```python
df['ano'] = df['Case Number'].str.replace('.*(\d{4}).*', r'\1', regex=True)
df['ano'].value_counts()
```

```python
mascara_ano = df['ano'].astype('float') > 1968
df_cinquenta_anos = df.loc[mascara_ano]
```

```python
evolucao = df_cinquenta_anos.groupby('ano')[['Case Number']].count().sort_values(by='ano', ascending=True)
evolucao_anual = evolucao.reset_index().rename(columns={'Case Number':'ataques'})
```

**Evolução dos ataques nos últimos 50 anos.**

```python
evolucao_anual
```
